```python
df=df.replace('McKail - Willyung','McKail')
df=df.replace('Gelorup - Dalyellup - Stratham','Gelorup')
df=df.replace('Lesmurdie - Bickley - Carmel','Lesmurdie')
df=df.replace('/n','')
df
```

```python
GP=[]
for i in Ciudad:
    latlng=geocoder.komoot(i + ', Australia').latlng
    GP.append(latlng)
print(GP) 
```

```python
lat1, lng1 = [], []
for n in GP:
    lat1.append(n[0])
    lng1.append(n[1])
    
print(lng1,lat1)
```

```python
lat1 = pd.Series(lat1)
lng1=pd.Series(lng1)
lat1 , lng1
```
